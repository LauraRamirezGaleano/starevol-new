```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
data_p1 = pd.read_csv('test_igws_on_0001062.p1',skiprows=2, header = None, sep= '\s+') 
data_p10 = pd.read_csv('test_igws_on_0001062.p10',skiprows=2, header = None, sep= '\s+') 

data_p1_15 = pd.read_csv('test_igws_on_0002462.p1',skiprows=2, header = None, sep= '\s+') 
data_p10_15 = pd.read_csv('test_igws_on_0002462.p10',skiprows=2, header = None, sep= '\s+') 
```

```python
# Assuming 'df' is your DataFrame
column_names_p1 = ["nsh", "yzi", "r", "T", "Rho", "P", "beta", "eta", "lnf", 
                "s", "Lr", "u", "xmr", "accel", "vr"]
column_names_p10 = ["nsh", "omega", "Ucirc", "Dshear", "Dcirc", "Dtot", "Dh", "psi", "xlambda", 
                "abmuj", "Nuturb", "Numol", "Kt", "aux", "SB", 'STh', 'SNG', 'SNS','Sadv', 'STh1', 'STh2', 'SB1', 'SB2', 'SB3', 'SB4', 'vomega' ]


# Assign column names
data_p1.columns = column_names_p1
data_p10.columns = column_names_p10

data_p1_15.columns = column_names_p1
data_p10_15.columns = column_names_p10
```

```python
plt.plot(data_p1['r'], data_p10['omega'], label = '12')
plt.plot(data_p1_15['r'], data_p10_15['omega'], label = '15')
plt.xlabel('Radius')
plt.ylabel('Omega')
plt.yscale('log')
plt.legend()
plt.show()
```
